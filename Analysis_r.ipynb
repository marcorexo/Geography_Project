```r
library(ggplot2)
library(reshape2)
```

```r
# Load the data
data_bangor <- read.csv("./bangor_transect.csv")
data_peris <- read.csv("./peris_transect.csv")
```

```r
str(data_bangor)
```

```r
data_bangor$Soil.Type <- NULL
data_peris$Soil.Type <- NULL
```

```r
# Melt the data into long format
melted_bangor_data <- melt(data_bangor)

# Create a box plot
ggplot(melted_bangor_data, aes(x = variable, y = value)) +
  geom_boxplot() +
  labs(title = "Box Plot for Data Frame",
       x = "",
       y = "")+
         theme(axis.text.x = element_text(angle = 90, vjust = 0.9, hjust=0.9), text = element_text(size = 25))
```

```r
# Melt the data into long format
melted_peris_data <- melt(data_peris)

# Create a box plot
ggplot(melted_peris_data, aes(x = variable, y = value)) +
  geom_boxplot() +
  labs(title = "Box Plot for Data Frame",
       x = "",
       y = "")+
         theme(axis.text.x = element_text(angle = 90, vjust = 0.9, hjust=0.9), text = element_text(size = 25))
```
